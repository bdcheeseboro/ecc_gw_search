```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
import os, glob
import numpy as np
import wquantiles as w
from matplotlib.backends.backend_pdf import PdfPages as pp
```

```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 1200
mpl.rcParams['figure.figsize'] = [5,4]
mpl.rcParams['text.usetex'] = True
mpl.rcParams["font.family"] = "serif"
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['axes.titlesize'] = 14
mpl.rcParams['xtick.labelsize'] = 8
mpl.rcParams['ytick.labelsize'] = 8
mpl.rcParams['legend.fontsize'] = 8
```

```python
mc_e0 = np.loadtxt('mc_e0_sorted.dat')
```

```python
fig, axs = plt.subplots(2,2)

fig.delaxes(axs[0][1])
axs[0,0].hist(mc_e0[:,0], density=True, bins=30, histtype='step');
axs[0,0].set(yticklabels=[])
axs[0,0].tick_params(left=False)
axs[0,1].set(yticklabels=[])
axs[0,1].set(xticklabels=[])
axs[1,0].hist2d(mc_e0[:,0], mc_e0[:,1], bins=30);
axs[1,0].set_xlabel(r'$e_{0}$')
axs[1,0].set_ylabel(r'$\log_{10}(\mathcal{M})$')
axs[1,1].hist(mc_e0[:,1], density=True, bins=30, histtype='step');
axs[1,1].set_xlabel(r'$\log_{10}(\mathcal{M})$')
axs[1,1].set(yticklabels=[])
axs[1,1].tick_params(left=False)
fig.tight_layout()
plt.savefig('mc_e0_corner.png')
```
